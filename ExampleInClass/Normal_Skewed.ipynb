```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Generate normal distribution
normal_data = np.random.normal(loc=10, scale=2, size=5000)

# Generate skewed distribution
skewed_raw = np.random.exponential(scale=1, size=5000)

# Rescale skewed distribution to match mean & variance
skewed_data = (
    (skewed_raw - skewed_raw.mean()) / skewed_raw.std()
) * normal_data.std() + normal_data.mean()

# Plot
plt.figure()
plt.hist(normal_data, bins=50, density=True, alpha=0.6)
plt.hist(skewed_data, bins=50, density=True, alpha=0.6)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Same Mean & Variance â€” Normal vs Skewed Distribution")
plt.legend(["Normal (Bell-shaped)", "Skewed"])

plt.show()

# Sanity check
print("Normal  mean, var:", normal_data.mean(), normal_data.var())
print("Skewed  mean, var:", skewed_data.mean(), skewed_data.var())
```
