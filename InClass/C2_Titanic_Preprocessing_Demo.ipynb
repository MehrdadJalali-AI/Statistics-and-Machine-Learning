## Data Preprocessing with the Titanic Dataset

This notebook demonstrates **basic data preprocessing steps** using the Titanic dataset.
It is a **simple in-class demo** to show how preprocessing works in practice.

## 1. Load Dataset

```python
import pandas as pd

# Load Titanic dataset (train.csv must be in the same folder)
df = pd.read_csv('train.csv')
df.head()
```

## 2. Select a Subset of Features

```python
features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']
target = 'Survived'

X = df[features].copy()
y = df[target]

X.head()
```

## 3. Handling Missing Values

```python
# Check missing values
X.isnull().sum()
```

```python
# Simple imputation
X['Age'] = X['Age'].fillna(X['Age'].median())
# fillna() is a pandas function used to replace missing values (NaN) in a dataset.
# df['Age'].fillna(30) // Every missing age becomes 30
X['Embarked'] = X['Embarked'].fillna(X['Embarked'].mode()[0])
#Takes the first most frequent value

X.isnull().sum()
```

## 4. Detecting and Handling Outliers

```python
from scipy.stats import zscore

# Z-score for Fare
fare_z = zscore(X['Fare'])
outliers = X[abs(fare_z) > 3]
# X[condition] selects rows of a DataFrame that satisfy a given condition.
outliers[['Fare']].head()
```

```python
# Remove extreme outliers
X = X[abs(fare_z) <= 3]
X.head()
```

## 5. Data Transformation (Encoding)

```python
# One-hot encoding for categorical features
X_encoded = pd.get_dummies(X, columns=['Sex'], drop_first=False)
X_encoded.head()
```

```python
pd.get_dummies(X, columns=['Embarked'], drop_first=False).head()
```

## 6. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_std = StandardScaler()
scaler_mm = MinMaxScaler()

X_scaled_std = scaler_std.fit_transform(X_encoded)
X_scaled_mm = scaler_mm.fit_transform(X_encoded)

X_scaled_std[:5], X_scaled_mm[:5]
```

## Summary
- Loaded a real-world dataset (Titanic)
- Handled missing values
- Detected and removed outliers
- Encoded categorical features
- Scaled numerical features

This prepares the data for machine learning models.
