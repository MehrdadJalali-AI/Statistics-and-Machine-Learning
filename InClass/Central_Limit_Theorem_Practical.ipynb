# Central Limit Theorem (CLT) â€“ Practical Demonstration

This notebook demonstrates the **Central Limit Theorem** using synthetic data.

**Key idea:**  
Even if the original population is **skewed**, the distribution of the **sample mean**
approaches a **normal distribution** as the sample size increases.



## Step 1: Import Libraries

- NumPy: data generation and statistics
- Matplotlib: visualization


```python

import numpy as np
import matplotlib.pyplot as plt
```


## Step 2: Set Random Seed

This ensures reproducibility.


```python

np.random.seed(42)
```


## Step 3: Generate a Skewed Population

We start with a **right-skewed population** using an exponential distribution.

**Why exponential?**
- Strongly skewed
- Clearly shows the power of CLT


```python

population_size = 100000
population = np.random.exponential(scale=50, size=population_size)
```


### Visualize the Population Distribution


```python

plt.figure()
plt.hist(population, bins=50)
plt.title("Original Population Distribution (Skewed)")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```


## Step 4: Sampling and Sample Means

We repeatedly:
1. Draw random samples from the population
2. Compute the **mean** of each sample

We do this for different sample sizes:
- n = 5
- n = 30
- n = 100


```python

def sample_means(population, sample_size, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        means.append(np.mean(sample))
    return np.array(means)
```


## Step 5: Distribution of Sample Means (n = 5)


```python

means_n5 = sample_means(population, sample_size=5)

plt.figure()
plt.hist(means_n5, bins=30)
plt.title("Distribution of Sample Means (n = 5)")
plt.xlabel("Sample Mean")
plt.ylabel("Count")
plt.show()
```


## Step 6: Distribution of Sample Means (n = 30)


```python

means_n30 = sample_means(population, sample_size=30)

plt.figure()
plt.hist(means_n30, bins=30)
plt.title("Distribution of Sample Means (n = 30)")
plt.xlabel("Sample Mean")
plt.ylabel("Count")
plt.show()
```


## Step 7: Distribution of Sample Means (n = 100)


```python

means_n100 = sample_means(population, sample_size=100)

plt.figure()
plt.hist(means_n100, bins=30)
plt.title("Distribution of Sample Means (n = 100)")
plt.xlabel("Sample Mean")
plt.ylabel("Count")
plt.show()
```


## Step 8: Interpretation

- The **original population** is strongly skewed
- For **small samples (n = 5)**, the distribution is still skewed
- For **larger samples (n = 30)**, the distribution becomes more symmetric
- For **large samples (n = 100)**, the distribution is approximately normal

This happens **regardless of the original population shape**.



## Final Takeaway

> The Central Limit Theorem explains why the normal distribution appears everywhere
in statistics and machine learning â€” especially when working with averages.
