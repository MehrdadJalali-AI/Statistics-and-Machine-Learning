# Practical Part (2): Measures of Dispersion

One notebook – two approaches (manual and package-based).

```python
data = [5, 7, 8, 7, 10, 6, 9, 8, 7, 6]
print('Data:', data)
```

## Task A – Without Python Packages
Pure Python implementation.

```python
# Range
range_manual = max(data) - min(data)

# Mean (needed for variance)
mean_manual = sum(data) / len(data)

# Variance
variance_manual = sum((x - mean_manual) ** 2 for x in data) / len(data)

# Standard Deviation
std_manual = variance_manual ** 0.5

range_manual, variance_manual, std_manual
```

## Task B – Using Python Packages
Using `statistics` and `numpy`.

```python
import statistics as stats
import numpy as np

range_pkg = max(data) - min(data)
variance_pkg = stats.pvariance(data)
std_pkg = stats.pstdev(data)

variance_np = np.var(data)
std_np = np.std(data)

range_pkg, variance_pkg, std_pkg, variance_np, std_np
```

## Task C – Interpretation

```python
print('Range:', range_manual)
print('Variance:', variance_manual)
print('Standard Deviation:', std_manual)

print('\nInterpretation:')
print('The data shows moderate spread around the mean, indicating reasonable consistency with some variability.')
```
