# Practical Part (3): Generating Synthetic Data with Distributions

This notebook demonstrates **step by step** how to generate **synthetic datasets** using different probability distributions,
visualize them, and analyze their statistical properties.

We will work with:
- Normal distribution
- Uniform distribution
- Poisson distribution

The goal is to understand **how distribution parameters influence data shape**.



## Step 1: Import Required Libraries

We use:
- **NumPy** for data generation
- **Matplotlib** for visualization
- **SciPy** for computing skewness


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
```


## Step 2: Set Random Seed

A random seed ensures **reproducibility**.
This means every run produces the same synthetic data.


```python

np.random.seed(42)
```


## Step 3: Generate Synthetic Data

### 1. Normal Distribution
- Mean (μ) = 50
- Standard deviation (σ) = 10
- Symmetric, bell-shaped

### 2. Uniform Distribution
- Values equally likely between 0 and 100
- Flat distribution

### 3. Poisson Distribution
- λ (rate) = 5
- Discrete, right-skewed


```python

n = 1000

normal_data = np.random.normal(loc=50, scale=10, size=n)
uniform_data = np.random.uniform(low=0, high=100, size=n)
poisson_data = np.random.poisson(lam=5, size=n)
```


## Step 4: Visualize the Distributions

We plot histograms to visually inspect:
- Shape
- Symmetry
- Skewness


```python

plt.figure()
plt.hist(normal_data, bins=30, alpha=0.7, density=True)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

plt.figure()
plt.hist(uniform_data, bins=30, alpha=0.7, density=True)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

plt.figure()
plt.hist(poisson_data, bins=30, alpha=0.7, density=True)
plt.title("Poisson Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```


## Step 5: Compute Summary Statistics

We compute:
- **Mean**: central tendency
- **Variance**: spread of data
- **Skewness**: asymmetry of distribution


```python

def summarize(data, name):
    print(f"{name}")
    print("Mean:", np.mean(data))
    print("Variance:", np.var(data))
    print("Skewness:", skew(data))
    print("-" * 30)

summarize(normal_data, "Normal Distribution")
summarize(uniform_data, "Uniform Distribution")
summarize(poisson_data, "Poisson Distribution")
```


## Step 6: Interpretation

### Observations
- **Normal distribution**: skewness close to 0 → most symmetric
- **Uniform distribution**: flat shape, skewness near 0
- **Poisson distribution**: positive skewness → right-skewed

### Key Insight
> Mean and variance alone do not describe distribution shape.  
> **Skewness captures asymmetry**, which is critical for ML models.



## Final Takeaway

Synthetic data helps us:
- Understand statistical assumptions
- Test ML models safely
- Visualize abstract concepts

Always inspect **distribution shape**, not just summary statistics.
