# Practical Part (1): Measures of Central Tendency

One notebook – two approaches.

```python
data = [5, 7, 8, 7, 10, 6, 9, 8, 7, 6]
print('Data:', data)
```

## Task A – Without Python Packages
Pure Python implementation.

```python
# Mean
mean_manual = sum(data) / len(data)

# Median
sorted_data = sorted(data)
n = len(sorted_data)
if n % 2 == 0:
    median_manual = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
else:
    median_manual = sorted_data[n//2]

# Mode
freq = {}
for value in data:
    freq[value] = freq.get(value, 0) + 1
mode_manual = max(freq, key=freq.get)

mean_manual, median_manual, mode_manual
```

## Task B – Using Python Packages
Using `statistics` and `numpy`.

```python
import statistics as stats
import numpy as np

mean_pkg = stats.mean(data)
median_pkg = stats.median(data)
mode_pkg = stats.mode(data)

mean_np = np.mean(data)
median_np = np.median(data)

mean_pkg, median_pkg, mode_pkg, mean_np, median_np
```

## Comparison

```python
print('Mean  → Manual:', mean_manual, '| statistics:', mean_pkg)
print('Median→ Manual:', median_manual, '| statistics:', median_pkg)
print('Mode  → Manual:', mode_manual, '| statistics:', mode_pkg)
```
