# Testing Differences Between Groups  
## t-Test and ANOVA (with Clear Interpretation for ML)

This notebook explains **how and why** we test differences between groups.

The focus is not only on *running the test*, but on **understanding the output**,
which is essential in **Machine Learning model comparison**.



## Learning Objectives

After completing this notebook, you should be able to:
- Explain why differences between groups can be misleading
- Interpret **t-statistic** and **p-value**
- Decide whether a difference is **real or random**
- Connect hypothesis testing to **ML model evaluation**



## Step 1: Import Libraries


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, f_oneway
```


## Step 2: Generate Synthetic Data (Two Groups)

We simulate two groups with slightly different averages.

Think of this as:
- Exam scores of two student groups  
- OR accuracy scores of two ML models


```python

np.random.seed(42)

group_A = np.random.normal(loc=70, scale=10, size=50)
group_B = np.random.normal(loc=75, scale=10, size=50)
```


## Step 3: Visualize the Groups


```python

plt.figure()
plt.hist(group_A, bins=20, alpha=0.6, label="Group A")
plt.hist(group_B, bins=20, alpha=0.6, label="Group B")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.title("Comparison of Two Groups")
plt.show()
```


## Step 4: t-Test (Comparing Two Groups)


```python

t_stat, p_value = ttest_ind(group_A, group_B)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```


## Understanding the t-Test Output

### t-statistic
- Measures how large the difference between means is
- Relative to variation inside the groups
- Larger absolute value → stronger difference

### p-value
- Probability of observing this difference by chance
- Small p-value → difference unlikely to be random

### Decision Rule
- p < 0.05 → significant difference
- p ≥ 0.05 → difference may be random



## Step 5: Generate Data for Three Groups


```python

group_C = np.random.normal(loc=80, scale=10, size=50)
```


## Step 6: Visualize Three Groups


```python

plt.figure()
plt.hist(group_A, bins=20, alpha=0.6, label="Group A")
plt.hist(group_B, bins=20, alpha=0.6, label="Group B")
plt.hist(group_C, bins=20, alpha=0.6, label="Group C")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.title("Comparison of Three Groups")
plt.show()
```


## Step 7: ANOVA (Three or More Groups)


```python

f_stat, p_value_anova = f_oneway(group_A, group_B, group_C)

print("F-statistic:", f_stat)
print("p-value:", p_value_anova)
```


## Understanding ANOVA

- F-statistic compares between-group vs within-group variation
- Small p-value → at least one group is different
- ANOVA does not say which group differs



## Final Takeaways

Statistical tests help us decide whether observed differences
are **real** or just due to **random variation**.

This is crucial when comparing ML models.
